```python
expr_a = expr_a.removeO()
if expr_a.has(sp.Sum):
    expr_a = replace_infinite_sums(expr_a)

expr_b = expr_b.removeO()
if expr_b.has(sp.Sum):
    expr_b = replace_infinite_sums(expr_b)

expr_a, expr_b
```

```python
import numpy as np
VAR_SUBSTITUTIONS = {
    A: lambda: 2 * np.random.randint(1, 10) + 1,
    B: lambda: 2 * np.random.randint(1, 10) + 1,
    D: lambda: 2 * np.random.randint(1, 10) + 1,
    E: lambda: 2 * np.random.randint(1, 10) + 1,
    F: lambda: 2 * np.random.randint(1, 10) + 1,
    x: lambda: 2 * np.random.randn()*100,
    sp.var('e'): lambda: sp.exp(1)
}
```

```python
diffs = []
for _ in range(n_attempts):
    subs = {}
    for symb in free_symbols:
        subs[symb] = VAR_SUBSTITUTIONS[symb]()

    numer_val_a = expr_a.subs(subs).doit()
    numer_val_b = expr_b.subs(subs).doit()
    diffs.append(abs(
        (numer_val_a - numer_val_b) / 
        (numer_val_a + numer_val_b)
    ))
    if diffs[-1] < allowed_diff:
        print(subs)
        break

diffs
```
